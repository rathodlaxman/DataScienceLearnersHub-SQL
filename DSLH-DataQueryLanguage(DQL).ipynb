### **Data Query Language (DQL)**

- In Microsoft SQL Server refers to the subset of SQL (Structured Query Language) that is specifically designed for querying and retrieving data from a database. DQL is one of the four main categories of SQL commands, and the other three categories are Data Definition Language (DDL), Data Manipulation Language (DML), and Data Control Language (DCL).
    
- DQL is primarily concerned with querying the database to retrieve information. The main SQL statement associated with DQL is the \`SELECT\` statement. The \`SELECT\` statement is used to retrieve data from one or more tables in a database based on specified criteria. Here is a basic syntax of the \`SELECT\` statement:
    

**Syntax :** 

SELECT column1, column2, ...

FROM table\_name

WHERE condition;

**Explanation :**

<span style="font-family: -apple-system, BlinkMacSystemFont, sans-serif; color: var(--vscode-foreground);">- </span> **SELECT Clause :**  

- Whenever u execute the SELECT statement the data goes to cache memory and from their the data is displayed. It gets created at the time of execution and removes after completion of query. So this columns are created at the runtime.
-  **\`column1, column2, ...\`**: The columns that you want to retrieve.

\- **FROM Clause:**

- The **FROM** clause is used to specify the table or tables from which the data should be retrieved. It is a mandatory part of the **SELECT** statement.
- **\`table\_name\`**: The table from which you want to retrieve data.

<span style="font-size: 13px;">-&nbsp;</span>   **WHERE Clause:**

- The **WHERE** clause is used to filter the rows returned by a **SELECT** statement based on a specified condition. It allows you to narrow down the result set to only include rows that meet specific criteria.
- <span style="color: rgb(74, 74, 74); font-family: -apple-system, &quot;system-ui&quot;, sans-serif; font-size: 14px;">Optional condition to filter the rows based on certain criteria.</span>

```sql
SELECT employee_id, first_name, last_name
FROM employees

-- Note : Here we have zero rows as until now we havent inserted any data in employees table.
-- We will insert data into employees table when we will be learning INSERT Command shortly
```

**Wildcard Character:**

- The asterisk (\*) can be used as a wildcard character to select all columns from a table.

```sql
SELECT *
FROM employees;
```
