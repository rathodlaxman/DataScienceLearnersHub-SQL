**SUM**: Calculates the sum of a numeric column.

```sql
SELECT SUM(Salary) AS total_salary_amount FROM employeeDetails
```

**AVG**: Calculates the average value of a numeric column.

```sql
SELECT AVG(Salary) AS average_salary_amount FROM employeeDetails
```

**MIN**: Returns the minimum value in a set of values.

**MAX**: Returns the maximum value in a set of values.

```sql
SELECT MIN(Salary) AS minimum_salary_amount FROM employeeDetails

SELECT MAX(Salary) AS maximum_salary_amount FROM employeeDetails
```

```sql
SELECT MIN(name) Min_with_char_value FROM employeeDetails

SELECT MAX(name) Max_with_char_value FROM employeeDetails
```

**COUNT(\*)**: This variant counts the total number of rows in a table, regardless of the values in any particular column. It counts all the rows, including those with NULL values.

```sql
SELECT COUNT(*) AS total_no_of_rows FROM employeeDetails
```

**COUNT(column\_name)**: This variant counts the number of non-NULL values in the specified column. It does not count rows where the specified column has a NULL value

```sql
SELECT COUNT(department) AS total_no_of_departments FROM employeeDetails

-- NOTE : Here in department column we dont have NULL values so all the rows are counted.
-- If there would have been two NULL values in departments column than the answer would have been
-- Total number of rows in department table - total number of null values in department column
```

**COUNT(DISTINCT column\_name)**: This variant counts the number of distinct (unique) non-NULL values in the specified column.

```sql
SELECT COUNT(DISTINCT department) AS total_no_of_departments FROM employeeDetails
```

**COUNT(1) or COUNT('literal') or COUNT(123)**: This variant counts the total number of rows in a table, similar to **COUNT(\*)**. The argument itself is not significant; it's just a placeholder.

```sql
SELECT COUNT(1) AS count_with_1 FROM employeeDetails
SELECT COUNT('some_string') AS count_with_some_string FROM employeeDetails
SELECT COUNT(123) AS count_with_random_number FROM employeeDetails
```

**STRING\_AGG**: Concatenates values from multiple rows into a single string, with an optional separator. Introduced in SQL Server 2017.

```sql
SELECT STRING_AGG(name, ', ') FROM employeeDetails;
```

```sql
SELECT
    Department,
    STRING_AGG(name, ', ') AS EmployeesInDepartment
FROM
    employeeDetails
GROUP BY
    Department;
```
