### **2\. UPDATE Statement**

- The **UPDATE** statement in MS SQL Server is used to modify existing records in a table. It allows you to change the values of one or more columns based on specified conditions.

<span style="font-size: 13px;"><b>Syntax :</b></span>

UPDATE table\_name

SET column1 = value1, column2 = value2, ...

WHERE condition;

**Explanation :**

- **table\_name**: The name of the table you want to update.
- **SET**: Specifies the columns and their new values.
- **column1 = value1, column2 = value2, ...**: The columns to be updated with their new values.
- **WHERE**: Specifies the conditions that must be met for the update to occur. **NOTE : If omitted, all rows in the table will be updated.**

```sql
USE DataScienceLearnersHub
```

```sql
SELECT * FROM EMP
```

**Updating a Single Column:**

```sql
UPDATE Emp
SET salary = 65000.00
WHERE id = 1;
```

```sql
SELECT * FROM EMP

-- NOTE : The value of salary of Rajesh gets updated from 60000 to 65000 as seen in the output
```

**Updating Multiple Column:**

```sql
UPDATE Emp
SET salary = 70000.00, department = 'Sales'
WHERE id = 2;
```

```sql
SELECT * FROM EMP
```

**Conditional Update:**

```sql
UPDATE Emp
SET salary = salary * 1.1
WHERE department = 'IT';
```

```sql
SELECT * FROM EMP
```

**Update All Rows:**

```sql
UPDATE Emp
SET age = age + 1;
```

```sql
SELECT * FROM EMP
```
