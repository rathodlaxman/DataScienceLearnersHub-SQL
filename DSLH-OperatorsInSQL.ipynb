#### Retrieve employees from the 'Sales' department with a salary greater than 6000.

```sql
SELECT * FROM employeeDetails
WHERE department = 'Sales' AND salary > 6000;
```

 **OR Operator:**

- Retrieve employees from the 'Finance' department or with a salary greater than 7000.

```sql
SELECT * FROM employeeDetails
WHERE department = 'Finance' OR salary > 7000;
```

**NOT Operator:**

- Retrieve employees not from the 'HR' department.

```sql
SELECT * FROM employeeDetails
WHERE NOT department = 'HR';
```

**Combining Logical Operators:**

- Retrieve female employees from the 'IT' department or with a salary less than 6000.

```sql
SELECT * FROM employeeDetails
WHERE (gender = 'Female' AND department = 'IT') OR salary < 6000;
```

**ALL Operator:**

- Retrieve employees with a salary greater than all salaries in the 'Marketing' department.

```sql
SELECT * FROM employeeDetails
WHERE salary > ALL (SELECT salary FROM employeeDetails WHERE department = 'Marketing');
```

**ANY Operator:**

- Retrieve employees with a salary greater than any salary in the 'IT' department.

```sql
SELECT * FROM employeeDetails
WHERE salary > ANY (SELECT salary FROM employeeDetails WHERE department = 'IT');
```

**EXISTS Operator:**

- Retrieve employees from the 'Finance' department where there exists a female employee.

```sql
SELECT * FROM employeeDetails
WHERE department = 'Finance' AND EXISTS (SELECT * FROM employeeDetails WHERE gender = 'Female');
```

**IN Operator:**

- Retrieve employees from the 'HR' and 'IT' departments.

```sql
SELECT * FROM employeeDetails
WHERE department IN ('HR', 'IT');
```

**LIKE Operator:**

- Retrieve employees with names starting with 'J'.

```sql
SELECT * FROM employeeDetails
WHERE name LIKE 'L%';
```

 **IS NULL Operator:**

- Retrieve employees with a null value in the 'gender' column.

```sql
SELECT * FROM employeeDetails
WHERE gender IS NULL;

-- NOTE : Here the table is empty because there are no NULL values in gender column
```

**NOT Operator:**

- Retrieve employees not from the 'Sales' department.

```sql
SELECT * FROM employeeDetails
WHERE department <> 'Sales';
```

 **UNIQUE Operator:**

- Retrieve unique department names from the table.

```sql
SELECT DISTINCT department FROM employeeDetails;
```

**BETWEEN Operator:**

- Retrieve employees with a salary between 6000 and 7000.

```sql
SELECT * FROM employeeDetails
WHERE salary BETWEEN 6000 AND 7000;
```
